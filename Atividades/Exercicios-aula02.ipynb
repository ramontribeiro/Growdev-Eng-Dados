### EXERCÍCIO 01: ✅



```python
print('--- Conversor de temperatura: °C - F ---');
temperatura_celsius = float(input('Digite a temperatura em °C: '));

temperatura_fahrenheit = ((9/5) * temperatura_celsius) + 32
print(f'A temperatura {temperatura_celsius} °C, equivale a {round(temperatura_fahrenheit, 2)} F.')
```

```python
print('--- Conversor de temperatura: F - °C ---');
temperatura_fahrenheit = float(input('Digite a temperatura em F: '));

temperatura_celsius = (temperatura_fahrenheit - 32) / 1.8
print(f'A temperatura {temperatura_fahrenheit} F, equivale a {round(temperatura_celsius, 2)} °C.')
```

### EXERCÍCIO 03 ✅

```python
print('--- OPERAÇÃO MATEMÁTICA ---');

print('''Escolha a operação
[1] Adição
[2] Subtração ''');

operacao = int(input('Informe a operação: '));
num1 = float(input('Número 01: '));
num2 = float(input('Número 02: '));

if operacao != 1 and operacao != 2:
  print('Operação inválida');
  operacao = int(input('Informe a operação: '));

if operacao == 1:
  print(f'Soma = {round(num1 + num2, 2)}')
elif operacao == 2:
  print(f'Subtração = {round(num1 - num2, 2)}')
else:
  print('Operação inválida - Excedeu tentativas');
```

### EXERCÍCIO 04 ✅


```python
print('--- SISTEMA DE NOTAS v1.0---');

nota1 = float(input('Nota 1° Bimestre: '));
nota2 = float(input('Nota 2° Bimestre: '));
nota3 = float(input('Nota 3° Bimestre: '));
nota4 = float(input('Nota 4° Bimestre: '));

total_aulas = int(input('Total de aulas: '));
total_faltas = int(input('Total de Faltas: '));

media = round((nota1 + nota2 + nota3 + nota4) / 4, 2);
faltas = round((total_faltas / total_aulas) * 100, 2);

if faltas > 25:
  print('Reprovado por Faltas');
elif media < 7:
  print('Reprovado por Média');
else:
  print('Aprovado');
```

```python
print('--- ATRIBUIÇÃO DE CONCEITO ---');

nota1 = float(input('Nota 1: '));
nota2 = float(input('Nota 2: '));

media = round((nota1 + nota2) / 2, 2);
conceito = '';
status = '';

if media < 4:
  conceito = 'E'
  status = 'REPROVADO'
elif media < 6:
  conceito = 'D'
  status = 'REPROVADO'
elif media < 7.5:
  conceito = 'C'
  status = 'REPROVADO'
elif media < 9:
  conceito = 'B'
  status = 'APROVADO'
else:
  conceito = 'A'
  status = 'APROVADO'

print('-------------');
print(f'Nota 1: {round(nota1, 2)}');
print(f'Nota 2: {round(nota2, 2)}');
print(f'Média: {media}');
print(f'Conceito: {conceito}');
print(status);
```

### EXERCÍCIO 09

```python
print('--- REAJUSTE SALARIAL ---');

salario = round(float(input('Qual o salário ($): ')), 2);
reajuste = 0;

if salario <= 280.00:
  reajuste = 0.2;
elif salario < 700.00:
  reajuste = 0.15;
elif salario < 1500.00:
  reajuste = 0.1;
else:
  reajuste = 0.05;

valor_reajuste = round(salario * reajuste, 2);
salario_reajuste = salario + valor_reajuste;

print('-------------');
print(f'Salário Atual: R$ {salario}');
print(f'Reajuste: {round((reajuste * 100), 2)} %');
print(f'Valor Aumento: R$ {valor_reajuste}');
print(f'Salário com Reajuste: R$ {salario_reajuste}');
```
